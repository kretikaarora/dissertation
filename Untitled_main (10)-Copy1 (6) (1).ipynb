```python
def to_number(value,suffix):
    value=float(value)
    if suffix=='k':
        return value*1e3
    elif suffix =='M':
        return value*1e6
    return value 
```

```python
def extract_investment_range(investment_str):
    if isinstance(investment_str, str):
        amounts = re.findall(r'€?([\d.]+)([kM]?)', investment_str.replace(',', ''))
        if len(amounts)==2:
           min_investment=to_number(*amounts[0]) 
           max_investment=to_number(*amounts[1])
           return pd.Series([min_investment,max_investment])
        elif len(amounts)==1:
            return pd.Series([to_number(*amounts[0]),None])
    return pd.Series([None,None])
```

```python
# ---------- 0) Ensure required columns exist ----------
for col in ["investment_areas", "funding_stage", "min_investment", "max_investment"]:
    if col not in df.columns:
        df[col] = np.nan

# ---------- 1) Stage normalization (run ONCE after loading df) ----------
def normalize_funding_stage(stage: str):
    if not isinstance(stage, str):
        return []
    s = stage.strip()
    # protect exact phrase
    s = s.replace("Series C & Beyond", "Series C and Beyond")
    s = s.replace("Series C and Beyond", "<<SERIES_C_AND_BEYOND>>")
    # split on '/', ',', ' and ', '&'
    parts = re.split(r'/|,| and |&', s, flags=re.IGNORECASE)
    out = []
    for p in parts:
        p = p.strip()
        if not p:
            continue
        if p == "<<SERIES_C_AND_BEYOND>>":
            out.append("series c and beyond")
        elif p.lower() == "b":
            out.append("series b")
        else:
            out.append(p.lower())
    return out

df["funding_stage_clean"] = df["funding_stage"].apply(normalize_funding_stage)
```

```python
import numpy as np 

def search_output(
    user_query,
    selected_stages=None,
    min_desired_investment=None,
    max_desired_investment=None,
    top_n=10,
    tfidf_weight=0.4,
    bert_weight=0.5,
    fuzzy_weight=0.1
):
    tfidf_scores = np.zeros(len(df))
    bert_scores = np.zeros(len(df))
    fuzzy_scores = np.zeros(len(df))

    if user_query:
        tfidf_user_vec = tfidf_vectorizer.transform([user_query.lower()])
        tfidf_partial = cosine_similarity(tfidf_user_vec, tfidf_matrix).flatten()
        tfidf_scores[mask_non_empty] = tfidf_partial

        bert_user_vec = bert_model.encode(user_query, convert_to_tensor=True)
        bert_partial = util.cos_sim(bert_user_vec, bert_embeddings).flatten().cpu().numpy()
        bert_scores[mask_non_empty] = bert_partial

        fuzzy_scores = df['investment_areas'].apply(lambda x: fuzz.partial_ratio(user_query.lower(), str(x)) / 100).values


    df['tfidf_score'] = tfidf_scores
    df['bert_score'] = bert_scores
    df['fuzzy_score'] = fuzzy_scores

    df['combined_score'] = (
        tfidf_weight * df['tfidf_score'] +
        bert_weight * df['bert_score'] +
        fuzzy_weight * df['fuzzy_score']
    )

    if selected_stages:    
        selected_stages = [s.lower() for s in selected_stages]
        stage_mask = df['funding_stage_clean'].apply(
            lambda stages: any(stage in stages for stage in selected_stages)
        )
        df_filtered = df[stage_mask]
    else:
        df_filtered = df
        
    if (
        min_desired_investment is not None and
        max_desired_investment is not None and
        min_desired_investment > max_desired_investment
    ):
        raise ValueError("Minimum desired investment cannot be greater than maximum desired investment.")


    if min_desired_investment is not None:
        df_filtered = df_filtered[
            df_filtered['max_investment'].isna() | (df_filtered['max_investment'] >= min_desired_investment)
        ]

    if max_desired_investment is not None:
        df_filtered = df_filtered[
            df_filtered['min_investment'].isna() | (df_filtered['min_investment'] <= max_desired_investment)
        ]

    results = df_filtered.sort_values(by='combined_score', ascending=False).head(top_n)
    
    return results[[
        'name', 'investment_areas', 'funding_stage', 'typical_investment',
        'min_investment', 'max_investment',
        'tfidf_score', 'bert_score', 'fuzzy_score', 'combined_score'
    ]]

# Example usage:
results = search_output(
    user_query="medical health",
    selected_stages=["series a", "series b", "series c and beyond"],
    min_desired_investment=0,
    max_desired_investment=200000,
    top_n=5
)

display(results)
```

```python
import re, math
from collections import Counter
import numpy as np
import pandas as pd


# ---------- 2) Shared filter helpers ----------
def filter_by_stage(view: pd.DataFrame, selected_stages):
    if not selected_stages or "funding_stage_clean" not in view.columns:
        return view
    wanted = {str(s).strip().lower() for s in selected_stages if isinstance(s, str)}
    mask = view["funding_stage_clean"].apply(
        lambda lst: any(isinstance(x, str) and x.strip().lower() in wanted for x in lst)
    )
    return view[mask]

def filter_by_investment(view: pd.DataFrame,
                         min_desired_investment: float,
                         max_desired_investment: float,
                         relax_if_empty: bool = True):
    if "min_investment" not in view.columns or "max_investment" not in view.columns:
        return view
    min_inv = pd.to_numeric(view["min_investment"], errors="coerce").fillna(0)
    max_inv = pd.to_numeric(view["max_investment"], errors="coerce").fillna(float("inf"))
    overlap = (min_inv <= max_desired_investment) & (max_inv >= min_desired_investment)
    kept = view[overlap]
    if kept.empty and relax_if_empty:
        return view.copy()
    return kept

def _minmax01(x: np.ndarray) -> np.ndarray:
    x = x.astype(float)
    mn, mx = float(np.min(x)), float(np.max(x))
    return (x - mn) / (mx - mn) if mx > mn else x

def _nonempty_mask(series: pd.Series) -> pd.Series:
    return series.notna() & series.astype(str).str.strip().ne("")

# ================== BM25 ==================
# ---------- 3) BM25 index (build ONCE) ----------
def _tok(s: str):
    return re.findall(r"[A-Za-z0-9]+", (s or "").lower())

mask_non_empty = _nonempty_mask(df["investment_areas"])
_bm25_docs = [_tok(t) for t in df.loc[mask_non_empty, "investment_areas"].fillna("")]
N = len(_bm25_docs)
_doc_len = np.array([len(d) for d in _bm25_docs], dtype=float)
_avgdl = _doc_len.mean() if N else 0.0

_dfreq = Counter()
for d in _bm25_docs:
    for w in set(d):
        _dfreq[w] += 1

_idf = {w: math.log(1 + (N - c + 0.5) / (c + 0.5)) for w, c in _dfreq.items()}
_k1, _b = 1.5, 0.75

# ---------- 4) BM25 query scorer ----------
def bm25_query_scores(user_query: str) -> np.ndarray:
    out = np.zeros(len(df), dtype=float)
    if not isinstance(user_query, str) or not user_query.strip():
        return out
    q = _tok(user_query)
    scores = np.zeros(N, dtype=float)
    for i, d in enumerate(_bm25_docs):
        if not d:
            continue
        denom_norm = _k1 * (1 - _b + _b * (_doc_len[i] / _avgdl)) if _avgdl > 0 else _k1
        tf = Counter(d)
        s = 0.0
        for w in q:
            f = tf.get(w, 0)
            if f == 0:
                continue
            s += _idf.get(w, 0.0) * (f * (_k1 + 1)) / (f + denom_norm)
        scores[i] = s
    out[mask_non_empty.values] = _minmax01(scores)
    return out

# ---------- 5) BM25 filtered search ----------
def bm25_search_output(user_query: str,
                       selected_stages=None,
                       min_desired_investment: float = 0,
                       max_desired_investment: float = float("inf"),
                       top_n: int = 20) -> pd.DataFrame:
    scores = bm25_query_scores(user_query)
    view = df.copy()
    view["bm25_score"] = scores
    view = filter_by_stage(view, selected_stages)
    view = filter_by_investment(view, min_desired_investment, max_desired_investment, relax_if_empty=True)
    view = view.sort_values("bm25_score", ascending=False)
    return view.head(top_n).reset_index(drop=True)

# ================== E5 (intfloat/e5-base-v2) ==================
# pip install sentence-transformers
from sentence_transformers import SentenceTransformer, util

# ---------- 6) E5 embeddings (build ONCE) ----------
_e5_model = SentenceTransformer("intfloat/e5-base-v2")
_e5_passages = ["passage: " + t for t in df.loc[mask_non_empty, "investment_areas"].fillna("")]
_e5_doc = _e5_model.encode(_e5_passages, normalize_embeddings=True, convert_to_tensor=True)

# ---------- 7) E5 query scorer ----------
def e5_query_scores(user_query: str) -> np.ndarray:
    out = np.zeros(len(df), dtype=float)
    if not isinstance(user_query, str) or not user_query.strip():
        return out
    q = _e5_model.encode(["query: " + user_query], normalize_embeddings=True, convert_to_tensor=True)
    sims = util.cos_sim(q, _e5_doc).cpu().numpy().ravel()
    sims = _minmax01(sims)
    out[mask_non_empty.values] = sims
    return out

# ---------- 8) E5 filtered search ----------
def e5_search_output(user_query: str,
                     selected_stages=None,
                     min_desired_investment: float = 0,
                     max_desired_investment: float = float("inf"),
                     top_n: int = 20) -> pd.DataFrame:
    scores = e5_query_scores(user_query)
    view = df.copy()
    view["e5_score"] = scores
    view = filter_by_stage(view, selected_stages)
    view = filter_by_investment(view, min_desired_investment, max_desired_investment, relax_if_empty=True)
    view = view.sort_values("e5_score", ascending=False)
    return view.head(top_n).reset_index(drop=True)

# ================== Example usage ==================
# (Adjust the range to your data—many VCs start above 200k.)
params = dict(
    user_query="digital health",
    selected_stages=["series a", "series b", "series c and beyond"],
    min_desired_investment=0,
    max_desired_investment=200000,
    top_n=5
)

res_bm25 = bm25_search_output(**params)
res_e5   = e5_search_output(**params)

# Show useful columns (won't error if some are missing)
bm25_cols = ["name","investment_areas","funding_stage","typical_investment","min_investment","max_investment","bm25_score"]
e5_cols   = ["name","investment_areas","funding_stage","typical_investment","min_investment","max_investment","e5_score"]

for c in bm25_cols:
    if c not in res_bm25.columns: res_bm25[c] = np.nan
for c in e5_cols:
    if c not in res_e5.columns: res_e5[c] = np.nan

display(res_bm25[bm25_cols])
display(res_e5[e5_cols])
```

```python

import os, re, numpy as np, pandas as pd
import compress_fasttext

# Download + load compressed model (only ~25MB)
FT_URL = "https://github.com/avidale/compress-fasttext/releases/download/v0.0.4/cc.en.300.compressed.bin"
FASTTEXT_BIN = os.getenv("FASTTEXT_BIN", "cc.en.300.compressed.bin")

if not os.path.exists(FASTTEXT_BIN):
    import urllib.request
    urllib.request.urlretrieve(FT_URL, FASTTEXT_BIN)

ft = compress_fasttext.models.CompressedFastTextKeyedVectors.load(FASTTEXT_BIN)

# ---------------- Helper functions ----------------
def _avg_vec_ft(text: str) -> np.ndarray:
    toks = re.findall(r"[A-Za-z]+", str(text).lower())
    if not toks:
        return np.zeros(ft.vector_size, dtype=float)
    vecs = [ft.get_vector(t) for t in toks if t in ft]
    if not vecs:
        return np.zeros(ft.vector_size, dtype=float)
    v = np.mean(np.vstack(vecs), axis=0)
    return v

# Build document matrix for non-empty rows
_ft_texts = df.loc[mask_non_empty, "investment_areas"].fillna("").astype(str).tolist()
_ft_doc = np.vstack([_avg_vec_ft(t) for t in _ft_texts])
_ft_doc = _ft_doc / (np.linalg.norm(_ft_doc, axis=1, keepdims=True) + 1e-12)

def fasttext_query_scores(user_query: str) -> np.ndarray:
    if not user_query or not str(user_query).strip():
        return np.zeros(len(df), dtype=float)
    q = _avg_vec_ft(user_query)
    q = q / (np.linalg.norm(q) + 1e-12)
    sim = q @ _ft_doc.T  # cosine sim (vectors already normalized)
    # normalize 0..1
    mn, mx = float(sim.min()), float(sim.max())
    sim = (sim - mn) / (mx - mn) if mx > mn else np.zeros_like(sim)
    out = np.zeros(len(df), dtype=float)
    out[mask_non_empty.values] = sim
    return out

def fasttext_search_output(user_query: str,
                           selected_stages=None,
                           min_desired_investment: float = 0.0,
                           max_desired_investment: float = 1e12,
                           top_n: int = 10):
    view = df.copy()
    view["fasttext_score"] = fasttext_query_scores(user_query)
    view = filter_by_stage(view, selected_stages)
    view = filter_by_investment(view, min_desired_investment, max_desired_investment)
    cols = ["name","investment_areas","funding_stage","typical_investment",
            "min_investment","max_investment","fasttext_score"]
    cols = [c for c in cols if c in view.columns]
    return view.sort_values("fasttext_score", ascending=False).head(top_n)[cols]

# ---------------- Example usage ----------------
results_fasttext = fasttext_search_output(
    user_query="digital health",
    selected_stages=["series a", "series b", "series c and beyond"],
    min_desired_investment=0,
    max_desired_investment=200000,
    top_n=5
)

display(results_fasttext[[
    "name","investment_areas","funding_stage",
    "typical_investment","min_investment","max_investment","fasttext_score"
]])
```

```python
# =================== HYBRIDS: BM25 + SBERT / BM25 + E5 ===================

import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer, util

# --- ensure helper(s) exist ---
if "_minmax01" not in globals():
    def _minmax01(arr: np.ndarray) -> np.ndarray:
        arr = np.asarray(arr, dtype=float)
        if arr.size == 0:
            return arr
        amin, amax = np.min(arr), np.max(arr)
        if amax > amin:
            return (arr - amin) / (amax - amin)
        return np.zeros_like(arr)

# --- ensure mask_non_empty exists (non-empty investment_areas) ---
if "mask_non_empty" not in globals():
    mask_non_empty = df["investment_areas"].notna() & df["investment_areas"].astype(str).str.strip().ne("")

# ---------------- SBERT (MiniLM) scorer ----------------
# Build SBERT embeddings once over non-empty rows
_sbert_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
_sbert_passages = df.loc[mask_non_empty, "investment_areas"].fillna("").tolist()
_sbert_doc = _sbert_model.encode(
    _sbert_passages, normalize_embeddings=True, convert_to_tensor=True
)

def sbert_query_scores(user_query: str) -> np.ndarray:
    """Return len(df) array of 0..1 scores for SBERT cosine similarity (min-max normalized per query)."""
    out = np.zeros(len(df), dtype=float)
    if not isinstance(user_query, str) or not user_query.strip():
        return out
    q = _sbert_model.encode([user_query], normalize_embeddings=True, convert_to_tensor=True)
    sims = util.cos_sim(q, _sbert_doc).cpu().numpy().ravel()  # cosine
    out[mask_non_empty.values] = _minmax01(sims)
    return out

def _sbert_query_raw(user_query: str) -> np.ndarray:
    """
    Return len(df) array of raw cosine similarities in [-1, 1] (zeros for empty rows).
    Used for Zhang-style candidate selection before per-query normalization on the union.
    """
    out = np.zeros(len(df), dtype=float)
    if not isinstance(user_query, str) or not user_query.strip():
        return out
    q = _sbert_model.encode([user_query], normalize_embeddings=True, convert_to_tensor=True)
    sims = util.cos_sim(q, _sbert_doc).cpu().numpy().ravel()
    out[mask_non_empty.values] = sims
    return out

# ---------------- Hybrid score combiner (kept for compatibility) ----------------
def hybrid_scores(bm25_arr: np.ndarray, dense_arr: np.ndarray, alpha: float = 0.6) -> np.ndarray:
    """
    Weighted sum: (1 - alpha) * BM25 + alpha * Dense
    Assumes both inputs are per-query normalized to [0,1].
    """
    alpha = float(alpha)
    return (1.0 - alpha) * bm25_arr + alpha * dense_arr

# =================== Zhang-style: parallel retrieval + union ===================

# Tunables for candidate pool sizes (feel free to adjust)
_K_BM25 = 500
_K_DENSE = 500

def _topk_indices(arr: np.ndarray, k: int) -> np.ndarray:
    """Return indices of the top-k values in arr (handles k > len)."""
    arr = np.asarray(arr)
    k = min(k, arr.size)
    if k <= 0 or arr.size == 0:
        return np.empty(0, dtype=int)
    idx = np.argpartition(arr, -k)[-k:]
    return idx[np.argsort(arr[idx])[::-1]]

def _normalize_on_subset(raw: np.ndarray, subset_idx: np.ndarray) -> np.ndarray:
    """
    Min-max normalize values only over subset_idx (per query).
    Returns a full-length array with zeros elsewhere.
    """
    out = np.zeros_like(raw, dtype=float)
    if subset_idx.size == 0:
        return out
    vals = raw[subset_idx].astype(float)
    vmin, vmax = vals.min(), vals.max()
    if vmax > vmin:
        vals = (vals - vmin) / (vmax - vmin)
    else:
        vals = np.zeros_like(vals)
    out[subset_idx] = vals
    return out

def _e5_query_raw(user_query: str) -> np.ndarray:
    """
    Raw E5 scores aligned to df. If you have precomputed E5 embeddings,
    mirror the SBERT pattern; otherwise fall back to existing e5_query_scores.
    """
    try:
        return e5_query_scores(user_query).astype(float)
    except NameError:
        return np.zeros(len(df), dtype=float)

# ---------------- Per-query hybrid scorers (arrays) ----------------
def hybrid_bm25_sbert_scores(user_query: str, alpha: float = 0.6) -> np.ndarray:
    """
    Zhang-style hybrid:
      1) Retrieve BM25 and SBERT in parallel
      2) Take union of top-K candidates
      3) Per-query normalize *on the union*
      4) Blend: (1 - alpha) * BM25 + alpha * SBERT
    Returns a len(df) array with blended scores (zeros outside the union).
    """
    # 1) Parallel retrieval (bm25_query_scores is expected to align to df)
    bm_raw = bm25_query_scores(user_query).astype(float)
    sb_raw = _sbert_query_raw(user_query).astype(float)

    # 2) Candidate union
    top_b = _topk_indices(bm_raw, _K_BM25)
    top_s = _topk_indices(sb_raw, _K_DENSE)
    if top_b.size == 0 and top_s.size == 0:
        return np.zeros(len(df), dtype=float)
    union_idx = np.unique(np.concatenate([top_b, top_s]))

    # 3) Per-query normalization on the union
    bm_norm = _normalize_on_subset(bm_raw, union_idx)
    sb_norm = _normalize_on_subset(sb_raw, union_idx)

    # 4) Blend
    alpha = float(alpha)
    hybrid = (1.0 - alpha) * bm_norm + alpha * sb_norm
    return hybrid

def hybrid_bm25_e5_scores(user_query: str, alpha: float = 0.6) -> np.ndarray:
    """
    Same as above but with E5 as the dense branch.
    """
    bm_raw = bm25_query_scores(user_query).astype(float)
    e5_raw = _e5_query_raw(user_query).astype(float)

    top_b = _topk_indices(bm_raw, _K_BM25)
    top_e = _topk_indices(e5_raw, _K_DENSE)
    if top_b.size == 0 and top_e.size == 0:
        return np.zeros(len(df), dtype=float)
    union_idx = np.unique(np.concatenate([top_b, top_e]))

    bm_norm = _normalize_on_subset(bm_raw, union_idx)
    e5_norm = _normalize_on_subset(e5_raw, union_idx)

    alpha = float(alpha)
    hybrid = (1.0 - alpha) * bm_norm + alpha * e5_norm
    return hybrid

# ---------------- Generic filtered search wrapper ----------------
def _hybrid_search_output(
    score_fn,                      # callable: user_query -> np.ndarray (len(df))
    user_query: str,
    selected_stages=None,
    min_desired_investment: float = 0,
    max_desired_investment: float = float("inf"),
    top_n: int = 20,
    score_col: str = "hybrid_score"
) -> pd.DataFrame:
    scores = score_fn(user_query)
    view = df.copy()
    view[score_col] = scores

    # reuse your shared filters
    view = filter_by_stage(view, selected_stages)
    view = filter_by_investment(view, min_desired_investment, max_desired_investment, relax_if_empty=True)

    view = view.sort_values(score_col, ascending=False)
    return view.head(top_n).reset_index(drop=True)

# ---------------- Public wrappers for each hybrid ----------------
def hybrid_bm25_sbert_search_output(
    user_query: str,
    selected_stages=None,
    min_desired_investment: float = 0,
    max_desired_investment: float = float("inf"),
    top_n: int = 20,
    alpha: float = 0.6
) -> pd.DataFrame:
    return _hybrid_search_output(
        score_fn=lambda q: hybrid_bm25_sbert_scores(q, alpha=alpha),
        user_query=user_query,
        selected_stages=selected_stages,
        min_desired_investment=min_desired_investment,
        max_desired_investment=max_desired_investment,
        top_n=top_n,
        score_col=f"hybrid_bm25_sbert_a{alpha:.2f}"
    )

def hybrid_bm25_e5_search_output(
    user_query: str,
    selected_stages=None,
    min_desired_investment: float = 0,
    max_desired_investment: float = float("inf"),
    top_n: int = 20,
    alpha: float = 0.6
) -> pd.DataFrame:
    return _hybrid_search_output(
        score_fn=lambda q: hybrid_bm25_e5_scores(q, alpha=alpha),
        user_query=user_query,
        selected_stages=selected_stages,
        min_desired_investment=min_desired_investment,
        max_desired_investment=max_desired_investment,
        top_n=top_n,
        score_col=f"hybrid_bm25_e5_a{alpha:.2f}"
    )

# ---------------- Optional: quick alpha sweep helper ----------------
def sweep_alpha_hybrid(user_query: str,
                       alphas=(0.2, 0.4, 0.5,0.6,0.7, 0.8),
                       which="bm25+e5",
                       selected_stages=None,
                       min_desired_investment: float = 0.0,
                       max_desired_investment: float = 1e12,
                       top_n: int = 10):
    out = {}
    for a in alphas:
        if which == "bm25+e5":
            out[a] = hybrid_bm25_e5_search_output(user_query, selected_stages,
                                                  min_desired_investment, max_desired_investment,
                                                  top_n, alpha=a)
        elif which == "bm25+sbert":
            out[a] = hybrid_bm25_sbert_search_output(user_query, selected_stages,
                                                     min_desired_investment, max_desired_investment,
                                                     top_n, alpha=a)
        else:
            raise ValueError("which must be 'bm25+e5' or 'bm25+sbert'")
    return out

# [Your existing hybrid function definitions...]
# hybrid_bm25_sbert_scores, hybrid_bm25_e5_scores
# hybrid_bm25_sbert_search_output, hybrid_bm25_e5_search_output

# =================== Example usage ===================
# (Your existing example usage code)

#
```

## Relevance Judgment Process

The `relevant_docs` dictionary was created through the following process:
1. Pooling: We took the top 20 results from each of the following models: BM25, SBERT, E5, USE, FastText, and the hybrid models.
2. Manual Curation: We then manually judged each of the pooled investors for each query to determine relevance.
3. This set of relevant documents is therefore a subset of the entire corpus. Metrics like Recall are calculated relative to this pool, not the absolute number of relevant documents in the entire corpus.

```python
# =========================
# EVALUATION SUITE (drop-in)
# =========================
import time, math, re, unicodedata
import numpy as np
import pandas as pd
from collections import Counter
from IPython.display import display

# ----------- Safety mask -----------
if "mask_non_empty" not in globals():
    # Fall back to common column names if needed
    _area_col_guess = "investment_areas" if "investment_areas" in df.columns else (
        "Investment Areas" if "Investment Areas" in df.columns else None
    )
    if _area_col_guess is None:
        raise ValueError(f"Could not find an Investment Areas column. Columns: {list(df.columns)}")
    mask_non_empty = df[_area_col_guess].notna() & df[_area_col_guess].astype(str).str.strip().ne("")

# =========================
# SIMPLE QRELS (no stage)  |
#  - robust preprocessing  |
#  - rich synonym map      |
#  - soft lexical matching |
# =========================
DOMAIN_SYNONYMS = {
    "ai": ["artificial intelligence","machine learning","ml","deep learning","neural networks",
           "genai","generative ai","foundation models","computer vision","nlp","natural language processing",
           "llm","large language model","autonomous agent","autonomous agents","predictive analytics","data science"],
    "fintech": ["financial technology","payments","payment processing","paytech","cards","issuing","acquiring",
                "banking","neobank","neo bank","open banking","personal finance","wealthtech","wealth tech",
                "trading","brokerage","capital markets","lending","credit","insurtech","regtech","defi",
                "stablecoin","crypto payments","remittance","cross border payments","bnpl","point of sale financing"],
    "healthtech": ["health tech","medtech","med tech","digital health","health care","healthcare","telemedicine",
                   "telehealth","ehealth","life sciences","biotech","pharma","clinical trial","clinical trials",
                   "drug discovery","diagnostic","diagnostics","medical device","medical devices","care delivery",
                   "mental health","femtech"],
    "biotech": ["synthetic biology","genomics","proteomics","cell therapy","gene therapy","crispr",
                "biomanufacturing","techbio","wet lab","lab automation","drug screening"],
    "climate tech": ["climatetech","cleantech","clean tech","greentech","green tech","renewables","renewable energy",
                     "solar","wind","geothermal","hydrogen","carbon capture","ccs","energy transition","sustainability",
                     "esg","impact investing","net zero","decarbonization","battery","batteries","energy storage","ev charging"],
    "iot": ["internet of things","industrial iot","iiot","connected devices","edge computing","embedded systems","sensors"],
    "saas": ["software as a service","enterprise software","b2b software","devtools","developer tools",
             "api platform","platform as a service","paas","cloud","multi tenant","multi-tenant","subscription"],
    "cybersecurity": ["security","infosec","appsec","cloud security","identity","iam","zero trust","siem","soar",
                      "threat detection","endpoint security","passwordless","pki","key management","data loss prevention","dlp"],
    "ecommerce": ["e commerce","e-commerce","online retail","marketplace","d2c","direct to consumer","omnichannel","shoptech","checkout"],
    "mobility": ["transportation","autonomous vehicle","autonomous vehicles","self driving","self-driving",
                 "ev","electric vehicle","electric vehicles","micro mobility","micromobility","fleet management","telematics"],
    "blockchain": ["web3","crypto","cryptocurrency","defi","dex","exchange","wallet","custody","tokenization",
                   "smart contracts","nft","on chain","on-chain","layer 1","layer 2","staking","zero knowledge","zk"],
    "data": ["analytics","big data","data platform","data engineering","data warehouse","business intelligence","bi",
             "data lake","data pipeline","feature store","mlops","model ops","etl","elt"],
    "ar vr": ["augmented reality","virtual reality","xr","metaverse","spatial computing"],
    "robotics": ["automation","robotic systems","hardware","robotics","cobots","industrial automation"],
    "proptech": ["real estate tech","real-estate","real estate","property technology","contech","construction tech",
                 "facility management","smart buildings","building management","bms"],
    "agritech": ["foodtech","food tech","agriculture technology","agri tech","precision agriculture","farmtech",
                 "alt protein","alternative protein","vertical farming","aquaculture"],
    "logistics": ["supply chain","warehouse","warehousing","last mile","last-mile","fulfilment","fulfillment",
                  "inventory management","freight","shipping","parcel","delivery","route optimization","3pl","4pl"],
    "space tech" : ["space","aerospace","satellite","earth observation","eo","gnss","navigation","launch","rocketry","orbital","ground segment"],
    "industry 4.0": ["manufacturing","additive manufacturing","3d printing","iiot","predictive maintenance","digital twin","plc","scada","mes"],
    "martech adtech": ["marketing technology","advertising technology","customer data platform","cdp","crm","campaign",
                       "attribution","ad network","dsp","ssp","programmatic","retail media"],
    "gaming xr": ["gaming","game dev","game development","game engine","esports","ar","vr","xr","metaverse","spatial computing"],
    "media": ["content","creator economy","ugc","streaming","ott","audio","podcast","video platform","news media"],
    "quantum": ["quantum computing","quantum sensing","quantum communication","qubits","qec","quantum algorithms"],
    "semiconductor": ["chips","chip design","asic","fpga","eda","foundry","packaging","advanced packaging","photolithography"],
    "telecom": ["5g","6g","ran","open ran","open-ran","telco cloud","network slicing","private networks"],
    "energy": ["power","grid","smart grid","demand response","virtual power plant","vpp",
               "distributed energy resources","derms","transmission","distribution","power electronics"],
}

# --- Normalization & variants ---
DELIM_CHARS = r"[\/&,+\-\–\—\(\)\[\]\{\}\|]"
NON_WORD = r"[^a-z0-9\s]"

def _norm(s: str) -> str:
    s = "" if s is None else str(s)
    s = unicodedata.normalize("NFKD", s.lower())
    s = re.sub(DELIM_CHARS, " ", s)       # break hyphens/slashes/etc.
    s = re.sub(NON_WORD, " ", s)          # keep [a-z0-9 ]
    s = re.sub(r"\s+", " ", s).strip()
    return s

def _plural_variants(term: str):
    out = {term}
    if term.endswith("y") and len(term) > 3: out.add(term[:-1] + "ies")
    if not term.endswith("s"): out.add(term + "s")
    if term.endswith("s") and len(term) > 1: out.add(term[:-1])
    return out

def _hyphen_variants(term: str):
    t = term.replace("-", " ").strip()
    parts = t.split()
    out = {t, "".join(parts)}
    if len(parts) > 1:
        out.add("-".join(parts))
    return out

def _expand_variants(term: str):
    base = _norm(term)
    variants = set()
    for v in _hyphen_variants(base):
        variants |= _plural_variants(v)
    return {_norm(v) for v in variants}

# --- Build label<->term maps once ---
def _build_term_maps(domain_syns: dict):
    label_to_terms, term_to_labels = {}, {}
    for label, terms in domain_syns.items():
        expanded = set()
        for t in set(terms) | {label}:
            expanded |= _expand_variants(t)
        lab = _norm(label)
        label_to_terms[lab] = sorted(expanded)
        for term in expanded:
            term_to_labels.setdefault(term, set()).add(lab)
    return label_to_terms, term_to_labels

_LABEL_TO_TERMS, _TERM_TO_LABELS = _build_term_maps(DOMAIN_SYNONYMS)

# --- Soft lexical matcher: whole-word OR token-overlap (Jaccard) for multi-word terms ---
def _soft_match(doc_txt: str, term: str, jacc_tau: float = 0.5) -> bool:
    dt = _norm(doc_txt)
    t  = _norm(term)
    if not dt or not t:
        return False
    if re.search(rf"\b{re.escape(t)}\b", dt):  # whole-word exact
        return True
    toks_t = set(t.split())
    if len(toks_t) >= 2:
        toks_d = set(dt.split())
        inter = len(toks_t & toks_d)
        jacc  = inter / (len(toks_t | toks_d) + 1e-9)
        return jacc >= jacc_tau
    return False

# --- Expand ANY query to the union of canonical label terms ---
def _expand_query_to_terms(q: str):
    qn = _norm(q)
    candidate_labels = set()
    # direct term→labels
    if qn in _TERM_TO_LABELS:
        candidate_labels |= _TERM_TO_LABELS[qn]
    # fuzzy fallback: token overlap against known terms
    q_tokens = set(qn.split())
    if not candidate_labels and q_tokens:
        for term, labels in _TERM_TO_LABELS.items():
            t_tokens = set(term.split())
            if not t_tokens: 
                continue
            inter = len(q_tokens & t_tokens)
            jacc  = inter / (len(q_tokens | t_tokens) + 1e-9)
            if jacc >= 0.5:
                candidate_labels |= labels
    # final fallback: just the query itself
    if not candidate_labels:
        return {qn}
    # union of terms across matched labels
    term_union = set()
    for lab in candidate_labels:
        term_union |= set(_LABEL_TO_TERMS.get(lab, []))
    return term_union

# --- (kept name) Build qrels with synonyms + soft matching, no stage gating ---
def build_qrels_simple_no_stage(queries, df, areas_col="investment_areas", jacc_tau=0.5):
    """
    Returns {query: set(df_indices)} using:
      - robust normalization + variants (hyphen/space/plural)
      - synonym expansions via DOMAIN_SYNONYMS
      - soft lexical matching (whole-word OR token-overlap)
    """
    # be resilient to column name variants
    if areas_col not in df.columns:
        for cand in ["Investment Areas", "InvestmentAreas", "Areas"]:
            if cand in df.columns:
                areas_col = cand
                break
    texts = df[areas_col].fillna("").astype(str).tolist()
    ids   = df.index.tolist()

    qrels = {}
    for q in queries:
        terms = sorted(_expand_query_to_terms(q))
        rel = set()
        for doc_id, txt in zip(ids, texts):
            if any(_soft_match(txt, t, jacc_tau=jacc_tau) for t in terms):
                rel.add(doc_id)
        qrels[q] = rel
    return qrels

# ----------- Metrics -----------
def precision_at_k(y_true_set, ranked_ids, k=10):
    hits = sum(1 for i in ranked_ids[:k] if i in y_true_set)
    return hits / max(k, 1)

def recall_at_k(y_true_set, ranked_ids, k=10):
    if not y_true_set: return 0.0
    hits = sum(1 for i in ranked_ids[:k] if i in y_true_set)
    return hits / len(y_true_set)

def average_precision(y_true_set, ranked_ids, k=10):
    """AP@K with denominator min(k, |R|)"""
    if not y_true_set:
        return 0.0
    ap, hits = 0.0, 0
    for i, doc_id in enumerate(ranked_ids[:k], start=1):
        if doc_id in y_true_set:
            hits += 1
            ap += hits / i
    denom = min(k, len(y_true_set))
    return ap / denom if denom > 0 else 0.0

def reciprocal_rank(y_true_set, ranked_ids, k=10):
    for i, doc_id in enumerate(ranked_ids[:k], start=1):
        if doc_id in y_true_set:
            return 1.0 / i
    return 0.0

def ndcg_at_k(y_true_set, ranked_ids, k=10):
    def dcg(scores):
        return sum((s / math.log2(i+2)) for i, s in enumerate(scores))
    gains = [1.0 if doc_id in y_true_set else 0.0 for doc_id in ranked_ids[:k]]
    ideal = sorted(gains, reverse=True)
    dcg_val = dcg(gains)
    idcg = dcg(ideal)
    return (dcg_val / idcg) if idcg > 0 else 0.0

# ----------- Ranking helpers -----------
def rank_top_k_from_scores(scores: np.ndarray, k=10):
    order = np.argsort(-scores)  # descending
    return order[:k].tolist(), scores[order[:k]]

def call_scorer_safe(scorer, user_query):
    try:
        return scorer(user_query)
    except Exception as e:
        print(f"[skip] {getattr(scorer,'__name__',str(scorer))}: {e}")
        return None

# ----------- Diversity & overlap -----------
def tokenise(s):
    return set(t.lower() for t in re.findall(r"[a-z0-9]+", str(s)))

def intra_list_similarity(top_idx, col=None):
    # Jaccard across token sets; lower is more diverse
    if not top_idx: return np.nan
    # default to whatever Areas column we used
    if col is None:
        col = "investment_areas" if "investment_areas" in df.columns else "Investment Areas"
    toks = [tokenise(df.loc[i, col]) for i in top_idx]
    if len(toks) == 1: return 0.0
    sims = []
    for i in range(len(toks)):
        for j in range(i+1, len(toks)):
            a, b = toks[i], toks[j]
            inter = len(a & b)
            union = len(a | b) or 1
            sims.append(inter/union)
    return float(np.mean(sims)) if sims else 0.0

def jaccard_overlap(a_idx, b_idx):
    a, b = set(a_idx), set(b_idx)
    u = len(a | b) or 1
    return len(a & b) / u

# =========================
# MODELS registry (auto-skip if absent)
# =========================
MODELS = []
if "tfidf_query_scores" in globals():      MODELS.append(("TFIDF", tfidf_query_scores))
if "bm25_query_scores" in globals():       MODELS.append(("BM25", bm25_query_scores))
if "sbert_query_scores" in globals():      MODELS.append(("SBERT", sbert_query_scores))
if "e5_query_scores" in globals():         MODELS.append(("E5", e5_query_scores))
if "use_query_scores" in globals():        MODELS.append(("USE", use_query_scores))
if "fasttext_query_scores" in globals():   MODELS.append(("fastText", fasttext_query_scores))
# Hybrids (if available)
if "hybrid_bm25_sbert_scores" in globals():
    MODELS.append(("Hybrid BM25+SBERT (a=0.6)", lambda q: hybrid_bm25_sbert_scores(q, alpha=0.6)))
if "hybrid_bm25_e5_scores" in globals():
    MODELS.append(("Hybrid BM25+E5 (a=0.6)",    lambda q: hybrid_bm25_e5_scores(q, alpha=0.6)))

# =========================
# Evaluation runner
# =========================
def evaluate_models(
    queries,
    qrels=None,
    k=10,
    selected_stages=None,
    min_desired_investment=0.0,
    max_desired_investment=1e12,
    restrict_to_filtered_corpus=False
):
    """
    If restrict_to_filtered_corpus=True, we evaluate on the subset of rows that pass the filters,
    and we also restrict qrels to that subset.
    """
    # Build filtered view (optional)
    view = df.copy()
    if selected_stages is not None and "filter_by_stage" in globals():
        view = filter_by_stage(view, selected_stages)
    if "filter_by_investment" in globals():
        view = filter_by_investment(view, min_desired_investment, max_desired_investment)
    filtered_index = set(view.index.tolist())

    # If no qrels provided, fall back to strict literal (rarely used now)
    if qrels is None:
        areas_col = "investment_areas" if "investment_areas" in df.columns else "Investment Areas"
        areas = df[areas_col].fillna("").astype(str).map(_norm)
        qrels = {}
        for q in queries:
            qn = _norm(q)
            rel_idx = areas[areas.str.contains(rf"\b{re.escape(qn)}\b", regex=True)].index.tolist()
            qrels[q] = set(rel_idx)

    # If restricting corpus, remap evaluation universe
    if restrict_to_filtered_corpus:
        qrels = {q: (set(idx) & filtered_index) for q, idx in qrels.items()}
        zero_rel = [q for q, s in qrels.items() if len(s) == 0]
        if zero_rel:
            print(f"[warn] {len(zero_rel)} queries have 0 relevant after filtering: {zero_rel[:5]}...")

    # Containers
    summary_rows, per_query_rows = [], []

    for model_name, scorer in MODELS:
        t0 = time.time()
        apL, rrL, recL, preL, ndcgL, iljL = [], [], [], [], [], []
        cov_hits = 0

        for q in queries:
            y_true_set = qrels.get(q, set())
            scores = call_scorer_safe(scorer, q)
            if scores is None:
                apL = []  # force skip
                break

            scores = np.asarray(scores, dtype=float)

            if restrict_to_filtered_corpus:
                not_allowed = list(set(range(len(df))) - filtered_index)
                if not_allowed:
                    scores[np.array(not_allowed, dtype=int)] = -1e9

            top_idx, _ = rank_top_k_from_scores(scores, k=k)

            ap   = average_precision(y_true_set, top_idx, k)
            rr   = reciprocal_rank(y_true_set, top_idx, k)
            rec  = recall_at_k(y_true_set, top_idx, k)
            pre  = precision_at_k(y_true_set, top_idx, k)
            ndcg = ndcg_at_k(y_true_set, top_idx, k)

            apL.append(ap); rrL.append(rr); recL.append(rec); preL.append(pre); ndcgL.append(ndcg)
            if any(i in y_true_set for i in top_idx): cov_hits += 1
            iljL.append(intra_list_similarity(top_idx))

            per_query_rows.append({
                "model": model_name,
                "query": q,
                "AP@%d" % k: ap,
                "RR@%d" % k: rr,
                "Recall@%d" % k: rec,
                "Precision@%d" % k: pre,
                "nDCG@%d" % k: ndcg,
                "intra_list_jaccard (↓=diverse)": iljL[-1],
                "relevant_in_corpus": len(y_true_set)
            })

        t1 = time.time()
        if not apL:
            print(f"[skip] {model_name}")
            continue

        summary_rows.append({
            "model": model_name,
            "MAP@%d" % k: float(np.mean(apL)),
            "MRR@%d" % k: float(np.mean(rrL)),
            "Recall@%d" % k: float(np.mean(recL)),
            "Precision@%d" % k: float(np.mean(preL)),
            "nDCG@%d" % k: float(np.mean(ndcgL)),
            "Coverage@%d" % k: cov_hits / max(len(queries), 1),
            "Avg Intra-list Jaccard (↓=diverse)": float(np.mean(iljL)),
            "Avg Latency (s/query)": (t1 - t0) / max(len(queries), 1)
        })

    eval_df = pd.DataFrame(summary_rows).sort_values(
        [f"MAP@{k}", f"MRR@{k}", f"Recall@{k}"], ascending=False
    )
    per_query_df = pd.DataFrame(per_query_rows)

    return eval_df, per_query_df

# =========================
# Overlap diagnostics
# =========================
def model_topk_overlap(queries, k=10):
    def topk_for_model(scorer):
        out = {}
        for q in queries:
            scores = call_scorer_safe(scorer, q)
            if scores is None: return None
            idx, _ = rank_top_k_from_scores(np.asarray(scores), k=k)
            out[q] = idx
        return out

    cache = {}
    for name, scorer in MODELS:
        topk = topk_for_model(scorer)
        if topk is not None:
            cache[name] = topk

    pairs = []
    names = list(cache.keys())
    for i in range(len(names)):
        for j in range(i+1, len(names)):
            a, b = names[i], names[j]
            jac = np.mean([jaccard_overlap(cache[a][q], cache[b][q]) for q in queries])
            pairs.append({"model_a": a, "model_b": b, f"Jaccard@{k}": jac})
    return pd.DataFrame(pairs).sort_values(f"Jaccard@{k}", ascending=False)

# =========================
# RUN: choose queries & filters
# =========================
QUERIES = ["fintech", "digital health", "AI", "proptech", "climate tech", "space tech"]

# Build improved QRELS (no stage gating here; stage is applied in evaluate_models via filters)
QRELS = build_qrels_simple_no_stage(QUERIES, df, jacc_tau=0.5)

# Small verification that synonyms are actually used (shows which terms fired)
_verify_rows = []
for q in QUERIES:
    terms_dbg = sorted(_expand_query_to_terms(q))[:12]  # preview terms used
    _verify_rows.append({"query": q, "example_terms_used": ", ".join(terms_dbg)})
display(pd.DataFrame(_verify_rows))

# Evaluate (only if your scorers exist; otherwise we just stop after building QRELS)
try:
    EVAL_PARAMS = dict(
        queries=QUERIES,
        qrels=QRELS,
        k=10,
        selected_stages=["seed", "series a", "pre-seed"],  # same as your setup
        min_desired_investment=0,
        max_desired_investment=200_000,
        restrict_to_filtered_corpus=True  # align with your UI filters
    )

    eval_df, per_query_df = evaluate_models(**EVAL_PARAMS)
    display(eval_df)
    display(per_query_df.sort_values(["query", "AP@10"], ascending=[True, False]).head(100))

    overlap_df = model_topk_overlap(QUERIES, k=10)
    display(overlap_df)
except Exception as e:
    print("Models not evaluated (likely because scorers aren't defined in this kernel).")
    print("Built synonym-aware QRELS successfully. Error during evaluation:", e)
# =========================
# END EVALUATION SUITE
# =========================
```

```python
# =========================================================
# END-CELL (add-only): Fig 1 (MAP@10 vs α), Table 1, t-tests
# Uses your existing metrics & hybrid functions
# =========================================================
import os, math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---------- 1) Convert QRELS (df-index sets) → relevant_docs (names) ----------
assert "df" in globals(), "df not found. Run data loading first."
assert "name" in df.columns, "df must have a 'name' column."

if "QRELS" in globals():
    relevant_docs = {q: df.loc[list(idxs), "name"].astype(str).tolist()
                     for q, idxs in QRELS.items()}
elif "relevant_docs" in globals():
    pass  # already available
elif "qrels_simple" in globals():
    relevant_docs = qrels_simple
elif "qrels" in globals():
    relevant_docs = qrels
else:
    raise NameError("No qrels found. Build QRELS with build_qrels_simple_no_stage(...) first.")

# ---------- 2) Config ----------
ALPHA_VALUES = [0.2, 0.4, 0.6, 0.8]
TOP_N = 10
E5_MODEL_NAME = "Hybrid BM25+E5"
SBERT_MODEL_NAME = "Hybrid BM25+SBERT"

def _have_sbert():
    return "hybrid_bm25_sbert_search_output" in globals()

# ---------- 3) Runners (use your existing hybrid functions) ----------
def _run_e5(query: str, alpha: float):
    return hybrid_bm25_e5_search_output(
        user_query=query,
        selected_stages=None,
        min_desired_investment=0,
        max_desired_investment=float("inf"),
        top_n=TOP_N,
        alpha=alpha,
    )

def _run_sbert(query: str, alpha: float):
    return hybrid_bm25_sbert_search_output(
        user_query=query,
        selected_stages=None,
        min_desired_investment=0,
        max_desired_investment=float("inf"),
        top_n=TOP_N,
        alpha=alpha,
    )

# ---------- 4) Evaluation (calls YOUR metric functions) ----------
def _evaluate_model(run_fn, alphas, top_n=10):
    rows, ap_per_query = [], {}
    for alpha in alphas:
        ap_list, p_list, r_list, ndcg_list = [], [], [], []
        for query, y_true_names in relevant_docs.items():
            res = run_fn(query, alpha=alpha)
            ranked_names = list(res["name"])
            y_true_set = set(y_true_names)

            # Use YOUR metric function names/signatures
            ap_list.append(average_precision(y_true_set, ranked_names, top_n))
            p_list.append(precision_at_k(y_true_set, ranked_names, top_n))
            r_list.append(recall_at_k(y_true_set, ranked_names, top_n))
            ndcg_list.append(ndcg_at_k(y_true_set, ranked_names, top_n))

        ap_per_query[alpha] = np.array(ap_list, dtype=float)
        rows.append({
            "alpha": alpha,
            "MAP@10": float(np.mean(ap_list)),
            "P@10": float(np.mean(p_list)),
            "R@10": float(np.mean(r_list)),
            "nDCG@10": float(np.mean(ndcg_list)),
        })
    return pd.DataFrame(rows).set_index("alpha").sort_index(), ap_per_query

# ---------- 5) Run evaluations (E5 + optional SBERT) ----------
all_tables, per_query_AP = [], {}

# E5
e5_table, e5_ap = _evaluate_model(_run_e5, ALPHA_VALUES, TOP_N)
e5_table.insert(0, "Model", E5_MODEL_NAME)
all_tables.append(e5_table.reset_index())
for a, arr in e5_ap.items():
    per_query_AP[(E5_MODEL_NAME, a)] = arr

# SBERT (optional)
if _have_sbert():
    sbert_table, sbert_ap = _evaluate_model(_run_sbert, ALPHA_VALUES, TOP_N)
    sbert_table.insert(0, "Model", SBERT_MODEL_NAME)
    all_tables.append(sbert_table.reset_index())
    for a, arr in sbert_ap.items():
        per_query_AP[(SBERT_MODEL_NAME, a)] = arr

# ---------- 6) Table 1: Comprehensive Alpha Evaluation ----------
table1 = pd.concat(all_tables, ignore_index=True)
metric_cols = ["MAP@10", "P@10", "R@10", "nDCG@10"]

def _bold_max(s):
    return ['font-weight: bold' if v == s.max() else '' for v in s]

display(
    table1.style
    .format({c: "{:.3f}" for c in metric_cols})
    .apply(_bold_max, subset="MAP@10")
    .apply(_bold_max, subset="P@10")
    .apply(_bold_max, subset="R@10")
    .apply(_bold_max, subset="nDCG@10")
    .hide(axis="index")
)
os.makedirs("figures", exist_ok=True)
table1.to_csv("figures/table1_alpha_eval.csv", index=False)



# ---------- 8) Statistical Analysis: paired t-tests on AP@10 ----------
try:
    from scipy.stats import t as student_t
except Exception:
    student_t = None

from math import sqrt

def _ttest_rel_safe(a, b):
    a = np.asarray(a, dtype=float); b = np.asarray(b, dtype=float)
    d = a - b
    df = len(d) - 1
    m = d.mean()
    sd = d.std(ddof=1)
    t = m / (sd / sqrt(len(d))) if sd > 0 else 0.0
    p = 2 * student_t.sf(abs(t), df) if student_t else None
    return t, p, df

def _stars(p):
    if p is None: return ""
    return "***" if p < 0.001 else "**" if p < 0.01 else "*" if p < 0.05 else ""

def _report_best(model_name, ap_dict):
    means = {a: float(np.mean(v)) for a, v in ap_dict.items()}
    best_a = max(means, key=means.get)
    second_a = max({a: m for a, m in means.items() if a != best_a},
                   key=lambda x: means[x]) if len(means) > 1 else None
    print(f"\n=== {model_name}: Paired t-test (AP@10) ===")
    for a in sorted(means): print(f"  α={a:.1f}: MAP@10={means[a]:.4f}")
    if second_a is None:
        print("  Not enough alpha values for a comparison.")
        return
    t, p, df = _ttest_rel_safe(ap_dict[best_a], ap_dict[second_a])
    print(f"  Best α={best_a:.1f} vs Runner-up α={second_a:.1f} → t({df})={t:.2f}, p={p:.3g} {_stars(p)}")

print("\n--- Statistical Analysis ---")
_report_best(E5_MODEL_NAME, e5_ap)
if _have_sbert():
    _report_best(SBERT_MODEL_NAME, sbert_ap)
```

```python
# 1) α-sweep: MAP@10 vs α (single)  OR  4-panel (MAP, P, R, nDCG) vs α
# - alpha_results_df columns expected: ["model","alpha","MAP@10", "P@10", "R@10", "nDCG@10"] (some may be missing; they'll be skipped)
# - models plotted: "Hybrid BM25+SBERT", "Hybrid BM25+E5"
# - highlights α=0.6 on the E5 curve
import os
import numpy as np
import matplotlib.pyplot as plt

def plot_alpha_sweep(alpha_results_df,
                     view="single",                    # "single" or "grid"
                     metrics=("MAP@10","P@10","R@10","nDCG@10"),
                     models=("Hybrid BM25+SBERT","Hybrid BM25+E5"),
                     highlight_alpha=0.6,
                     save_dir="figures",
                     filename_single="figure1_alpha_curve.png",
                     filename_grid="figure_alpha_grid.png"):
    """
    view='single' → one plot: MAP@10 vs α (both hybrids)
    view='grid'   → 2x2 subplots: MAP@10, P@10, R@10, nDCG@10 vs α (both hybrids)
    """

    if alpha_results_df is None or alpha_results_df.empty:
        print("[plot_alpha_sweep] Nothing to plot.")
        return

    # ensure output dir
    os.makedirs(save_dir, exist_ok=True)

    # Helper: plot one metric on given axes
    def _plot_metric(ax, metric_name):
        any_line = False
        for name_prefix in models:
            sub = alpha_results_df[alpha_results_df["model"].str.startswith(name_prefix)]
            if sub.empty or metric_name not in sub.columns:
                continue
            agg = (sub.groupby("alpha", as_index=False)[metric_name]
                       .mean()
                       .sort_values("alpha"))
            ax.plot(agg["alpha"], agg[metric_name], marker="o", label=name_prefix)
            any_line = True

            # Highlight α for E5
            if name_prefix == "Hybrid BM25+E5" and highlight_alpha in agg["alpha"].values:
                y_opt = float(agg.loc[agg["alpha"] == highlight_alpha, metric_name].iloc[0])
                ax.scatter([highlight_alpha], [y_opt], s=120, edgecolor="black", zorder=3)
                ax.annotate(f"α={highlight_alpha:.1f}\n{metric_name}={y_opt:.3f}",
                            (highlight_alpha, y_opt),
                            xytext=(highlight_alpha+0.05, y_opt),
                            arrowprops=dict(arrowstyle="->", lw=1))
        ax.set_xlabel("Alpha (α)")
        ax.set_ylabel(metric_name)
        ax.grid(True, alpha=0.3)
        if any_line:
            ax.legend()

    if view == "single":
        # Single-metric: MAP@10
        fig, ax = plt.subplots(figsize=(7,5))
        _plot_metric(ax, "MAP@10")
        ax.set_title("Hybrid α-sweep (MAP@10)")
        plt.tight_layout()
        outpath = os.path.join(save_dir, filename_single)
        plt.savefig(outpath, dpi=300, bbox_inches="tight")
        plt.show()
        print("\nFigure 1 Caption: Performance of hybrid BM25+E5 model across different alpha values. Optimal performance achieved at α=0.6.")
        print(f"[saved] {outpath}")
    else:
        # Grid view: up to 4 metrics (skip missing ones)
        wanted = [m for m in metrics if m in alpha_results_df.columns]
        if not wanted:
            print("[plot_alpha_sweep] No requested metrics found in alpha_results_df.")
            return
        # Pad/truncate to 4 slots for a 2x2 grid
        grid_metrics = (wanted + ["", "", "", ""])[:4]

        fig, axes = plt.subplots(2, 2, figsize=(11, 8))
        axes = axes.flatten()
        for ax, metric_name in zip(axes, grid_metrics):
            if not metric_name:
                ax.axis("off")
                continue
            _plot_metric(ax, metric_name)
            ax.set_title(f"Hybrid α-sweep ({metric_name})")

        plt.tight_layout()
        outpath = os.path.join(save_dir, filename_grid)
        plt.savefig(outpath, dpi=300, bbox_inches="tight")
        plt.show()
        print("[saved]", outpath)
        print("Grid figure shows trade-offs across MAP@10, P@10, R@10, and nDCG@10 vs α (where available).")
```
